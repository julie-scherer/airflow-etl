```python
import re
unique = set()
for row in df.amenities:
    # print(f"row = {type(row)}, {row}")
    
    new = row.strip("[]")
    # print(f"new = {type(new)}, {new}")

    if "\\u" in new:
        new = new.replace("\\u"," ")
    if "\\" in new:
        new = new.replace("\\"," ")
    
    new_split = re.split('["]+|^[, ]$', new)
    # print(f"new_split = {type(new_split)}, {new_split}")

    clean = list(filter(lambda x: x != "" and x != ", ", new_split))
    # print(f"clean = {type(clean)}, {clean}")

    unique.update(set(clean))

unique
```
